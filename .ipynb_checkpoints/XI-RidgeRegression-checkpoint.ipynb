# Three types of Regularizations - 
1) Ridge Regression - L2 Regularization (We take Lambda*Msquare)bcause of m square we call it L2  - To avoid overfitting in the data. 
- Overfitting - model learns pattern on the training data too well along with the noise that it performs good on training data ver well but falls apart on the test or unseen data. In overfitting model the value of weights or coefficents or slope is too high. 
- IN the line y = mx+b
- If slope m = 0 then y is not dependent on x. Hence this is the case of under fitting
- if slope m = almost infinity then small change in x, the y changes drastically. This is the case of over fitting. So to avoid over fitting we need to reduce the slope m.
Model has low bias but high variance.
2) Lasso Regression - L1 Regularization
3) ElasticNet Combination of both Ridge & Lasso

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
from sklearn.datasets import load_diabetes

data = load_diabetes()
```

```python
print(data.DESCR)
```

```python
X = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=101)
```

# Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
L = LinearRegression()
```

```python
L.fit(X_train, y_train)
```

```python
y_pred = L.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

# Using Ridge Regression

```python
from sklearn.linear_model import Ridge, ridge_regression
R = Ridge(alpha = 0.0001)
#r = ridge_regression(X_train, y_train,alpha = 0.0001)
# alpha is a lambda or the hyper parameter
```

```python
R.fit(X_train, y_train)
```

```python
y_pred1 = R.predict(X_test)
```

```python
r2_score(y_test, y_pred1)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred1))
```

```python
m = 100
x1 = 5 * np.random.rand(m, 1) - 2
x2 = 0.7 * x1 ** 2 - 2 * x1 + 3 + np.random.randn(m, 1)

plt.scatter(x1, x2)
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

def get_preds_ridge(x1, x2, alpha):
    model = Pipeline([
        ('poly_feats', PolynomialFeatures(degree=16)),
        ('ridge', Ridge(alpha=alpha))
    ])
    model.fit(x1, x2)
    return model.predict(x1)

alphas = [0, 20, 200]
cs = ['r', 'g', 'b']

plt.figure(figsize=(10, 6))
plt.plot(x1, x2, 'b+', label='Datapoints')

for alpha, c in zip(alphas, cs):
    preds = get_preds_ridge(x1, x2, alpha)
    # Plot
    plt.plot(sorted(x1[:, 0]), preds[np.argsort(x1[:, 0])], c, label='Alpha: {}'.format(alpha))

plt.legend()
plt.show()
```

```python
# Redline is overfitting
# Blue line is underfitting
# Green line is generalizing
```
